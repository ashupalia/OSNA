```python
from collections import Counter
import matplotlib.pyplot as plt
import networkx as nx
import sys
import time
from TwitterAPI import TwitterAPI
from nltk.tokenize import word_tokenize
import pandas as pd
import networkx as networkx
import matplotlib.pyplot as plt
import numpy as np

###################################################################################################
#~~~~~~~~~~~~~~~This function returns a graph, and also creates an output png file~~~~~~~~~~~~~~~~#
###################################################################################################

def create_graph(quor = 0):
    data = list()
    edges = list()
    data.append('Chicago-cluster.csv')
    data.append('London-cluster.csv')
    data.append('Sydney-cluster.csv')
    graph=nx.Graph()
    for f in data:
        frame=pd.read_csv(f,sep='\t')
        for index,row in frame.iterrows():
            l=list(row['friends'].split(','))
            for x in l:
                edges.append((row['Id'],x))
    
    graph.add_edges_from(edges)
    nx.draw(graph)
    plt.savefig("cluster.png", format = "PNG")
    n=str(nx.nodes(graph))
    return graph

#################################################################################################################
#~~~~~~~~~~~~~~~This function uses girvan_newman to find the betweeness and to identify clusters~~~~~~~~~~~~~~~~#
#################################################################################################################
def girvan_newman(depth=0, min_nodes = 10, max_nodes =800, max_n = 0, index = 0):
    
    G=create_graph()
    if G.order() == 1:
        return [G.nodes()]
    result_components=[]
    file1=open("cluster.txt","w+")
    
   
    component = [c for c in nx.connected_component_subgraphs(G)]
    
    i=1;
    if(len(component)>1):
        for idx,l in enumerate(component):
            if max_n<len(l.nodes()):
                max_n=len(l.nodes())
                index=idx
        #index=largest(component)
    else:
        index=0

    indent = '   ' * depth  
    len_components_nodes = str(len(component[index].nodes()))
    print("Selected cluster with "+len_components_nodes)
    components=[component[index]]
    for i,c in enumerate(component):
        if i!=index:
            result_components.append(c)
    while len(components) == 1:
        g=nx.Graph(components[0])
        eb = nx.edge_betweenness_centrality(g)
        edge_to_remove = sorted(eb.items(), key=lambda x: x[1], reverse=True)
        edge =edge_to_remove[0][0]
        print(indent + 'removing ' + str(edge))
        g.remove_edge(*edge)
        comp = [c for c in nx.connected_component_subgraphs(g)]
        com = len(comp)
        if( com <=1 ):
            components=[g]
        else:
            components=comp
            
            
    result_components.extend(components)
    res_comp = len(result_components)
    print("Resultant components="+str(res_comp))
    if file1.mode=="w+":
        res_com = len(result_components)
        file1.write("\n Total number of communities discovered="+str(res_com))
        file1.write("\n Average number of user per community ="+str(len(G.nodes())/len(result_components)))
        file1.write("\n")
        file1.close()
    
    return components

def main():
    girvan_newman()

if __name__ == '__main__':
    main()
```
